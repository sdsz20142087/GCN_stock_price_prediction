```python
import numpy as np
import pandas as pd
```

```python
#导入share_holding_graph数据
data=pd.read_excel('share_holding_newsest_tongyi.xlsx')
data
```

```python
#导入股价graph
all_stock = pd.read_excel('df_csi300_total_five_year_f.xlsx',index_col=0)
all_stock
```

```python
#拿到CSI300的所有stock名称
stock_a=[]
for i in range(299):
    stock_a.append(all_stock.columns[i])
stock_a
```

```python
#拿到有股权记录的CSI300stock
data2 = data[data['证券代码'].isin(stock_a)]
data2
```

```python
set(stock_a).difference(set(list(data[data['证券代码'].isin(stock_a)].iloc[:,0])))
```

```python
#所有sharehold_name
sharehold_name = []
for i in range(10):
    for j in range(len(data2)):
        sharehold_name.append(data2.iloc[j,i+2])
```

```python
#去重去空后所有sharehold_name
sharehold_name2=[]
for i in set(sharehold_name):
    if type(i)!=float:
        sharehold_name2.append(i)
```

```python
#所有share_holding的持股数据
graph_data = []
for i in sharehold_name2:
    name_ratio=[]
    for k in range(10):
        for j in range(len(data2)):
            if i == data2.iloc[j,k+2]:
                name_ratio.append([data2.iloc[j,0],data2.iloc[j,k+12]])
    graph_data.append([i,name_ratio])
```

```python
graph_data
```

```python
#拿到多个公司共同拥有一个投资人的情况
graph_data_based=[]
for i in range(len(graph_data)):
    if len(graph_data[i][1]) >1:
        graph_data_based.append(graph_data[i])
```

```python
graph_data_based
```

```python
graph_data_based2=graph_data_based.copy()
```

```python
#拿到数字表示的邻接矩阵格式
for i in range(len(graph_data_based2)):
    for j in range(len(graph_data_based2[i][1])):
        for k in range(len(stock_a)):
            if stock_a[k] == graph_data_based2[i][1][j][0]:
                graph_data_based2[i][1][j][0]=k
graph_data_based2
```

```python
#计算出stock之间的相关性强度
k_row=[]
two_company_share=[]
for i in range(len(graph_data_based2)):
    one_line_data=[]
    one_sharehold=[]
    for j in range(len(graph_data_based2[i][1])):
        one_line_data.append(graph_data_based2[i][1][j][0])
        one_sharehold.append(graph_data_based2[i][1][j][1])
    #print(one_line_data)
    #print(one_sharehold)
    for f in range(len(one_line_data)-1):
        for k in range(len(one_line_data)-1-f):
            #print([one_line_data[f],one_line_data[f+k+1],((one_sharehold[f]+one_sharehold[f+k+1])/200)])
            k_row.append([one_line_data[f],one_line_data[f+k+1],((one_sharehold[f]+one_sharehold[f+k+1])/200)])
```

```python
k_row
```

```python
stock_relation = np.zeros((len(stock_a),len(stock_a)))
stock_relation.shape
```

```python
#构建出share_holding_graph邻接矩阵
for i in range(len(k_row)):
    stock_relation[k_row[i][0],k_row[i][1]]+=k_row[i][2]
    stock_relation[k_row[i][1],k_row[i][0]]+=k_row[i][2]
```
