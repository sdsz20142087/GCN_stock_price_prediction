```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_excel('/Users/yingzhedong/Desktop/图神经网络合集/df_csi300_total_five_year_f.xlsx',index_col=0)
data
```

```python
stock_value = pd.read_excel('/Users/yingzhedong/Desktop/图神经网络合集/市值数据.xlsx')
stock_value
```

```python
industry_level = pd.read_excel('/Users/yingzhedong/Desktop/图神经网络合集/industry_all.xlsx')
industry_level
```

```python
new_industry_level = industry_level[industry_level['证券代码'].isin(data.columns[:299])]
new_industry_level
```

```python
new_industry_stock_value = stock_value[stock_value['证券代码'].isin(data.columns[:299])]
new_industry_stock_value
```

```python
new_industry_stock_value.to_excel('stock_index_name.xlsx')
```

```python
#申万二级
stock_level_value = []
for i in range(len(new_industry_stock_value)):
    stock_level_value.append([new_industry_stock_value.iloc[i,0],new_industry_stock_value.iloc[i,2],new_industry_level.iloc[i,7]])
```

```python
#可匹配的有297支stock
data_name = list(data.columns[:299])
data_name
```

```python
for i in range(len(stock_level_value)):
    for j in range(len(data_name)):
        if stock_level_value[i][0] == data_name[j]:
            stock_level_value[i][0] = j
```

```python
#转换成编号
stock_level_value
```

```python
industry_relation_graph = np.zeros((len(data_name),len(data_name)))
industry_relation_graph.shape
```

```python
#此(N,N)矩阵的i行j列所对应的值为Mi/Mj
for i in range(len(stock_level_value)):
    for j in range(len(stock_level_value)):
        if stock_level_value[i][2] == stock_level_value[j][2]:
            industry_relation_graph[i,j] = stock_level_value[i][1]/stock_level_value[j][1]
industry_relation_graph
```

```python
np.save('industry_relation_graph.npy',industry_relation_graph)
```
