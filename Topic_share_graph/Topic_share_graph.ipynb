```python
import numpy as np
import pandas as pd
```

```python
data_topic = pd.read_excel('/Users/yingzhedong/Desktop/图神经网络合集/tonghuashun_news_2019-2020.xlsx',index_col=0)
data_topic
```

```python
data_stock_name = pd.read_excel('/Users/yingzhedong/Desktop/图神经网络合集/stock_index_name.xlsx',index_col=0)
data_stock_name
```

```python
stock_name = list(data_stock_name['证券简称'])
stock_name
```

```python
#定义共同文本识别函数
def text_recognize(name,content):
    relation = []
    flag=False
    for i in range(len(name)):
        if type(content)!=float:
            if name[i] in content:
                relation.append(i)
                flag=True
    if flag==False:
        relation.append(-1)
    return relation
```

```python
#形成关系列表
data_topic['relations'] = data_topic['content'].apply(lambda x :text_recognize(stock_name,x))
data_topic['relations']
```

```python

sum_rel = []
for i in range(len(data_topic['relations'])):
    if len(data_topic['relations'].iloc[i])>1:
        sum_rel.append(data_topic['relations'].iloc[i])
```

```python
sum_rel
```

```python
#形成stock纬度关系列表
relation_metrix = np.zeros((299,299))
for i in range(len(sum_rel)):
    len_k = len(sum_rel[i])
    for k in range(len_k):
        for j in range(k+1,len_k):
            relation_metrix[sum_rel[i][k],sum_rel[i][j]]+=1
            relation_metrix[sum_rel[i][j],sum_rel[i][k]]+=1
relation_metrix
```

```python
np.save('stock_topic_share_graph.npy',relation_metrix)
```
