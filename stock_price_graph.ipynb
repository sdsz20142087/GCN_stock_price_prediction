```python
#列名的截取
cut_stock_name=[]
for i in range(len(data.columns)):
    cut_stock_name.append(data.columns[i][:9])
```

```python
#独立的股票代号
set_cut_stock_name = set(cut_stock_name)
len(set_cut_stock_name)
```

```python
#构建input矩阵
stock_matrix = np.zeros((len(data),len(set_cut_stock_name),6))
stock_matrix
```

```python
#矩阵shape
stock_matrix.shape
```

```python
set_cut_stock_name = cut_stock_name[0:299]
```

```python
len(set_cut_stock_name)
```

```python
for i in range(len(data)):
    print(i/len(data))
    for j in range(len(set_cut_stock_name)):
        flag=0
        for k in range(data.shape[1]):
            if set_cut_stock_name[j] in data.columns[k]:
                stock_matrix[i,j,flag] = data.iloc[i,k]
                flag+=1
stock_matrix
```
